 # **Análisis Descriptivo**

Un análisis descriptivo es una parte fundamental de la estadística y la investigación científica. Se refiere al proceso de resumir, organizar y presentar datos de manera significativa para comprender sus características básicas. El objetivo principal del análisis descriptivo es describir y resumir los datos, en lugar de hacer inferencias o generalizaciones sobre una población más amplia.

1. **Resumen de datos:** Se resume la información mediante medidas estadísticas como la media, la mediana, la moda, la desviación estándar, etc.

2. **Visualización de datos:** Se utilizan gráficos y diagramas para representar los datos de manera visual, lo que facilita la comprensión de patrones, tendencias y distribuciones.

3. **Exploración de datos:** Se examinan los datos para identificar posibles valores atípicos, patrones interesantes, relaciones entre variables, etc.

4. **Organización de datos:** Los datos se organizan de manera que sean más fáciles de entender y analizar, como tablas, histogramas, diagramas de dispersión, entre otros.

5. **Presentación de resultados:** Los resultados del análisis descriptivo suelen presentarse de manera clara y concisa, utilizando tablas, gráficos y resúmenes estadísticos.

## **3. Distribucion de Frecuencias**

La distribución de frecuencias es una herramienta básica en estadística descriptiva que resume y organiza los datos mostrando cuántas veces ocurre cada valor en un conjunto de datos. Consiste en tabular los datos en clases o categorías y contar la frecuencia con la que aparece cada una de estas clases. Es fundamental en el análisis descriptivo por varias razones:

1. **Resumen de datos:** La distribución de frecuencias proporciona un resumen claro y conciso de la distribución de los datos. Ayuda a entender la forma y la estructura de los datos de manera rápida y eficiente.

2. **Identificación de patrones:** Permite identificar patrones y tendencias en los datos, como modas (valores más frecuentes), sesgos, dispersión, concentraciones, entre otros. Estos patrones pueden dar pistas sobre el comportamiento y las características de la población subyacente.

3. **Detectar valores atípicos:** Al observar la distribución de frecuencias, es posible identificar valores extremos o atípicos que pueden requerir una investigación adicional. Estos valores pueden influir en el análisis y en las conclusiones obtenidas a partir de los datos.

4. **Comparación entre grupos:** La distribución de frecuencias puede utilizarse para comparar la distribución de datos entre diferentes grupos o subpoblaciones. Esto puede ayudar a identificar diferencias significativas entre grupos y a comprender mejor las variaciones dentro de la población.

5. **Planificación de análisis posteriores:** La comprensión de la distribución de frecuencias puede guiar la selección de técnicas de análisis posteriores. Por ejemplo, puede ayudar a determinar qué pruebas estadísticas son más apropiadas o qué técnicas de visualización serían más efectivas para explorar más a fondo los datos.

#### Metodo 1:

```python
# Realizamos primero el conteo o frecuencias de la variable 'Sexo'
sexo_frecuencia = datos['Sexo'].value_counts()

#datos['Sexo']': Seleccionamos la columna 'Sexo' del DataFrame 'datos'.
#.value_counts()': Aplicamos la función 'value_counts()' para contar la frecuencia de cada valor único en la columna 'Sexo'.
#sexo_frecuencia': Almacenamos los resultados en una variable llamada 'sexo_frecuencia'.
```

```python
# Calculamos los porcentajes de las frecuencias de la variable 'Sexo'
# Se normaliza para obtener los porcentajes en lugar de las frecuencias absolutas
# Se multiplica por 100 para obtener los porcentajes y se redondea a dos decimales

sexo_porcentaje = (datos['Sexo'].value_counts(normalize=True) * 100).round(2)

# datos['Sexo']': Seleccionamos la columna 'Sexo' del DataFrame 'datos'.
# .value_counts(normalize=True)': Aplicamos la función 'value_counts()' con el parámetro 'normalize=True' para obtener las frecuencias normalizadas.
# * 100': Multiplicamos por 100 para obtener los porcentajes.
# .round(2)': Redondeamos los porcentajes a dos decimales.
# sexo_porcentaje': Almacenamos los resultados en una variable llamada 'sexo_porcentaje'.
```

```python
# Creamos el DataFrame 'dist_frec_sexo' con las frecuencias y los porcentajes de la variable 'Sexo'

dist_frec_sexo = pd.DataFrame({'Frecuencia': sexo_frecuencia, 'Porcentaje (%)': sexo_porcentaje})

# pd.DataFrame(): Creamos un DataFrame utilizando el constructor de pandas.
# {Frecuencia: sexo_frecuencia, Porcentaje (%):sexo_porcentaje}: Pasamos un diccionario donde las claves son
#   los nombres de las columnas y los valores son las series de datos (frecuencias y porcentajes) que hemos calculado anteriormente.
# dist_frec_sexo: Asignamos este DataFrame a la variable dist_frec_sexo para su posterior uso.
```

```python
# Renombrar los índices del DataFrame
dist_frec_sexo.rename(index={0: 'Masculino', 1: 'Femenino'}, inplace=True)

# 'index={0: 'Masculino', 1: 'Femenino'}': Renombra los índices 0 y 1 a 'Masculino' y 'Femenino' respectivamente.
# 'inplace=True': Realiza el cambio directamente en el DataFrame 'dist_frec_sexo'.

# Renombrar el nombre del eje de las columnas del DataFrame
dist_frec_sexo.rename_axis('Sexo', axis='columns', inplace=True)

# 'rename_axis('Sexo', axis='columns')': Renombra el eje de las columnas del DataFrame como 'Sexo'.
# 'axis='columns'': Indica que se está renombrando el eje de las columnas.
# 'inplace=True': Realiza el cambio directamente en el DataFrame 'dist_frec_sexo'.
```

Las variables cuantitativas también pueden ser analizadas siempre y cuando estén previamente categorizadas según algún criterio o clase personalizada que hayamos definido. Esto nos permite agrupar los datos en categorías significativas que faciliten su análisis y comprensión.

#### Clases personalisadas

##### Ingreso

Para analizar el comportamiento de la variable **INGRESO**, vamos a elaborar una tabla de frecuencias dividiendo los ingresos en distintas clases según el salario mínimo (SM) como referencia.

A continuación se detallan las clases de ingreso:

<b>A</b> ► Más de 20 SM

<b>B</b> ► De 10 a 20 SM

<b>C</b> ► De 4 a 10 SM

<b>D</b> ► De 2 a 4 SM

<b>E</b> ► Hasta 2 SM

Para establecer estas clases de ingreso, consideramos que el salario mínimo era de $788,000 miles de pesos colombianos.

Una vez construida la tabla, se solicita describir los puntos más relevantes observados tanto en la tabla como en el gráfico resultante.


* Definir los intevalos de las clases

```python
# Definición de clases para la variable de ingreso, basadas en múltiplos del salario mínimo ($788,00 miles de pesos colombianos)
clases = [
    datos.Ingreso.min(),  # Límite inferior de la primera clase, igual al ingreso mínimo del conjunto de datos
    2 * 788,              # Límite superior de la primera clase, establecido en 2 veces el salario mínimo
    4 * 788,              # Límite superior de la segunda clase, establecido en 4 veces el salario mínimo
    10 * 788,             # Límite superior de la tercera clase, establecido en 10 veces el salario mínimo
    20 * 788,             # Límite superior de la cuarta clase, establecido en 20 veces el salario mínimo
    datos.Ingreso.max()   # Límite superior de la última clase, igual al ingreso máximo del conjunto de datos
]
```

* Definir los labels de las clases

```python
# Definición de las etiquetas de las clases para la variable de ingreso, en orden inverso al proporcionado en la lista de clases
labels = ['E', 'D', 'C', 'B', 'A']
```

* Construir la columna de frecuencias:

    La función **cut()** de pandas se utiliza para segmentar y categorizar los valores de una variable en intervalos específicos, lo que permite agrupar los datos en clases discretas. Aquí tienes una explicación detallada de esta función:

    * **x:** Es la variable que se va a segmentar y categorizar.
    * **bins:** Son los límites de los intervalos en los que se van a dividir los datos. Pueden ser una secuencia de números que definen los límites de los intervalos, o un entero que indica la cantidad de intervalos que se desean crear.
    * **labels:** Son las etiquetas que se asignan a cada intervalo. Si no se proporcionan, la función devuelve solo los índices de los intervalos.
    * **include_lowest:** Es un booleano que indica si el límite inferior del primer intervalo debe estar incluido o no. Por defecto, es False.

    La función **cut()** devuelve una serie de datos cuyos valores son las categorías asignadas a cada observación en función de los intervalos especificados en bins. Estas categorías pueden ser utilizadas posteriormente para análisis estadísticos, visualizaciones o agrupaciones de datos.

```python
# Calcular la frecuencia de los valores categorizados de la variable de ingreso utilizando los límites de clase definidos en la lista 'clases' y las etiquetas definidas en la lista 'labels'
frecuencia = pd.value_counts(
    pd.cut(x=datos.Ingreso,  # Variable de ingreso a categorizar
           bins=clases,      # Límites de clase definidos anteriormente
           labels=labels,    # Etiquetas de clase definidas anteriormente
           include_lowest=True  # Incluir el límite inferior de la primera clase
          )
)
frecuencia
```

* Construir la columna de porcentajes

```python
# Calcular el porcentaje de frecuencia de los valores categorizados de la variable de ingreso utilizando los límites de clase definidos en la lista 'clases' y las etiquetas definidas en la lista 'labels'
porcentaje = (pd.value_counts(
    pd.cut(
        x=datos.Ingreso,        # Variable de ingreso a categorizar
        bins=clases,            # Límites de clase definidos anteriormente
        labels=labels,          # Etiquetas de clase definidas anteriormente
        include_lowest=True     # Incluir el límite inferior de la primera clase
    ),
    normalize=True  # Normalizar para obtener porcentajes
) * 100).round(2)
porcentaje
```

* Juntar las columnas de frecuencias y porcentajes y ordenar las líneas según los labels de las clases

```python
# Crear un DataFrame llamado 'dist_frec_ingreso' que contiene las frecuencias y los porcentajes de los valores categorizados de la variable de ingreso
dist_frec_ingreso = pd.DataFrame({'Frecuencia': frecuencia, 'Porcentaje (%)': porcentaje})

# Ordenar el DataFrame por el índice (las etiquetas de las clases) en orden descendente
dist_frec_ingreso.sort_index(ascending=False)
```

#### Clases de tamaño fijo

##### Ingreso

Ahora vamos a explorar la construcción de distribuciones de frecuencias para datos cuantitativos cuando no tenemos información previa sobre cómo categorizar los datos.

Utilizaremos la regla de Sturges para determinar el número óptimo de clases.

La regla de **Sturges** es una fórmula empírica utilizada para determinar el número óptimo de clases en una distribución de frecuencias para datos cuantitativos, cuando no se dispone de información previa sobre la categorización de los datos. La fórmula se basa en el tamaño de la muestra y se calcula de la siguiente manera:



formula.png

Donde ***n*** es el tamaño de la muestra. Esta fórmula sugiere que el número de clases óptimo aumenta con el tamaño de la muestra, lo que permite capturar adecuadamente la variabilidad de los datos sin generar demasiadas o muy pocas clases. La regla de Sturges es ampliamente utilizada en estadística para la construcción de histogramas y otras representaciones gráficas de datos cuantitativos.

* Numero optimo de clases (Regla de Sturges)

```python
# Calcular el tamaño de la muestra
n = datos.shape[0]  # datos.shape[0] devuelve el número de filas en el DataFrame 'datos'
```

```python
# Calcular el número óptimo de clases utilizando la fórmula de la regla de Sturges
k = 1 + (10 / 3) * np.log10(n)  # Donde 'n' es el tamaño de la muestra
```

```python
# Redondear el número óptimo de clases y convertirlo en un entero
k = int(k.round(0))
```

* Crear la tabla de frecuencias

```python
# Calcular la frecuencia de los valores categorizados de la variable de ingreso utilizando el método 'cut' de Pandas
frecuencia = pd.value_counts(
    pd.cut(
        x=datos.Ingreso,       # Variable de ingreso a categorizar
        bins=k,                # Número óptimo de clases calculado previamente
        include_lowest=True    # Incluir el límite inferior del primer intervalo
    ),
    sort=False                # No ordenar las clases
)
```

### 4.2 Distribución de frecuencia: Asimetrica

Una distribución de frecuencia asimétrica es un tipo de distribución en la que los valores se distribuyen de manera desigual alrededor de un punto central, lo que resulta en una forma de campana que no es simétrica cuando se representa gráficamente. En este tipo de distribución, la media, la mediana y la moda pueden estar ubicadas en diferentes puntos a lo largo del eje horizontal.

Existen dos tipos principales de distribuciones de frecuencia asimétricas:

1. **Asimetría a la derecha (sesgo positivo):** En esta situación, la cola de la distribución se extiende hacia la derecha, lo que significa que hay valores atípicos o extremos en el extremo derecho del rango de valores. En este caso, la media suele ser mayor que la mediana y la moda.

Las distribuciones de frecuencia asimétricas son comunes en la práctica y pueden surgir por una variedad de razones, incluidas desviaciones de la norma, factores subyacentes complejos o sesgos en los datos de muestreo. Es importante tener en cuenta la asimetría al analizar y interpretar datos, ya que puede afectar la interpretación de estadísticas descriptivas y la aplicación de métodos estadísticos.

## **5. Medidas de tendencia central**

Las medidas de tendencia central son estadísticas que representan un valor típico o central en un conjunto de datos. Son útiles para resumir la distribución de los datos en una sola cantidad. Las tres medidas de tendencia central más comunes son:

1. **Media (Promedio):** La media es la suma de todos los valores en un conjunto de datos dividida por el número total de observaciones. Es sensible a los valores extremos y puede verse afectada por ellos.

2. **Mediana:** La mediana es el valor que se encuentra en el centro de un conjunto de datos ordenados de manera ascendente. Divide el conjunto de datos en dos partes iguales, con el 50% de los datos por encima y el 50% por debajo. La mediana es menos sensible a los valores extremos que la media y es una medida de tendencia central más robusta en presencia de valores atípicos.

3. **Moda:** La moda es el valor que ocurre con mayor frecuencia en un conjunto de datos. Puede haber una moda (unimodal), dos modas (bimodal) o más (multimodal) en un conjunto de datos. La moda es útil para describir distribuciones de datos discretos o cuando se busca el valor más común en un conjunto de datos.

Estas medidas de tendencia central proporcionan diferentes perspectivas sobre la distribución de los datos y se utilizan en una variedad de contextos para resumir y comprender la naturaleza de los conjuntos de datos.

### **5.4 Relacion entre media, mediana y moda**

La relación entre la media, la mediana y la moda está relacionada con la distribución de los datos en un conjunto de datos. Estas tres medidas estadísticas de tendencia central proporcionan información sobre la ubicación central de los datos, pero pueden ser diferentes dependiendo de la forma de la distribución de los datos.

1. **Distribución sesgada a la derecha:** Cuando la mayoría de los datos están concentrados en el extremo derecho del eje de coordenadas, la moda es generalmente menor que la mediana, y la mediana es generalmente menor que la media. Esto se debe a que los valores extremadamente altos "estiran" la distribución hacia la derecha, aumentando la media más que la mediana y la moda.

2. **Distribución simétrica:** Cuando los datos están distribuidos simétricamente alrededor de un valor central, la media, la mediana y la moda son aproximadamente iguales. Por ejemplo, en una distribución normal (también conocida como distribución gaussiana), la media, la mediana y la moda son todas iguales.

3. **Distribución sesgada a la izquierda:** Cuando la mayoría de los datos están concentrados en el extremo izquierdo del eje de coordenadas, la moda es generalmente mayor que la mediana, y la mediana es generalmente mayor que la media. Esto se debe a que los valores extremadamente bajos "estiran" la distribución hacia la izquierda, disminuyendo la media más que la mediana y la moda.

## **6. Medidas de localizacion**

Los cuartiles, deciles y percentiles son medidas adicionales de tendencia central y dispersión que dividen un conjunto de datos ordenados en partes iguales o proporcionales.

1. **Cuartiles:** Los cuartiles dividen un conjunto de datos en cuatro partes iguales. El primer cuartil (Q1) es el valor que deja el 25% de los datos por debajo de él, la mediana (Q2) es el segundo cuartil, que deja el 50% de los datos por debajo y el tercer cuartil (Q3) deja el 75% de los datos por debajo.

2. **Deciles:** Los deciles dividen un conjunto de datos en diez partes iguales. El primer decil (D1) deja el 10% de los datos por debajo de él, el segundo decil (D2) deja el 20% de los datos por debajo y así sucesivamente hasta el décimo decil (D10), que deja el 90% de los datos por debajo.

3. **Percentiles:** Los percentiles dividen un conjunto de datos en cien partes iguales. El percentil p deja el p% de los datos por debajo de él. Por ejemplo, el percentil 25 (P25) es el primer cuartil (Q1), el percentil 50 (P50) es la mediana y el percentil 75 (P75) es el tercer cuartil (Q3).

Estas medidas proporcionan información detallada sobre la distribución y la posición relativa de los datos en un conjunto de datos. Son especialmente útiles para identificar valores atípicos, comprender la variabilidad de los datos y comparar diferentes conjuntos de datos en términos de su distribución.

### **6.1 Cuartiles**

Los cuartiles son medidas estadísticas que dividen un conjunto de datos ordenados en cuatro partes iguales. Estas partes son conocidas como el primer cuartil (Q1), la mediana (Q2) y el tercer cuartil (Q3). La mediana es el valor que deja el 50% de los datos por debajo de él, mientras que Q1 deja el 25% de los datos por debajo y Q3 deja el 75% de los datos por debajo.

Para calcular los cuartiles, primero ordenamos los datos de menor a mayor (o viceversa). Luego, encontramos la posición de cada cuartil utilizando la fórmula:

* Para Q1: (n + 1) / 4
* Para Q2 (la mediana): (n + 1) / 2
* Para Q3: 3 * (n + 1) / 4

Donde "n" es el número total de datos en el conjunto.

Si el resultado de estas fórmulas no es un número entero, se puede tomar el promedio de los valores en las posiciones inmediatamente superiores e inferiores para obtener el cuartil correspondiente.

Por ejemplo, si tenemos los siguientes datos: 12, 15, 20, 25, 30, 35, 40, 45, 50, 55, 60.

El primer cuartil (Q1) estaría en la posición (11 + 1) / 4 = 3, es decir, en el tercer dato (20).
La mediana (Q2) estaría en la posición (11 + 1) / 2 = 6, es decir, en el sexto dato (35).
El tercer cuartil (Q3) estaría en la posición 3 * (11 + 1) / 4 = 9, es decir, en el noveno dato (55).
Entonces, en este conjunto de datos, Q1 = 20, Q2 = 35 y Q3 = 55. Estos valores dividen el conjunto de datos en cuatro partes iguales en términos de número de observaciones.

```python
datos.Ingreso.quantile([.25,.5,.75])
```

### **6.2 Deciles**

Los deciles son medidas estadísticas que dividen un conjunto de datos ordenados en diez partes iguales. Estas partes son conocidas como D1, D2, ..., D9. El primer decil (D1) deja el 10% de los datos por debajo de él, el segundo decil (D2) deja el 20% de los datos por debajo, y así sucesivamente, hasta el noveno decil (D9), que deja el 90% de los datos por debajo.

Para calcular los deciles, primero ordenamos los datos de menor a mayor (o viceversa). Luego, encontramos la posición de cada decil utilizando la fórmula:

* Para Dk: (k/10) * (n + 1)

Donde "k" es el número de decil que estamos buscando (por ejemplo, k = 1 para D1, k = 2 para D2, etc.), y "n" es el número total de datos en el conjunto.

Si el resultado de esta fórmula no es un número entero, se puede tomar el promedio de los valores en las posiciones inmediatamente superiores e inferiores para obtener el decil correspondiente.

Por ejemplo, si tenemos los siguientes datos: 10, 15, 20, 25, 30, 35, 40, 45, 50, 55, 60.

1. El primer decil (D1) estaría en la posición (1/10) * (11 + 1) = 1.2, es decir, entre el primer y el segundo dato. Podemos tomar el promedio de estos dos valores para obtener D1.
2. El segundo decil (D2) estaría en la posición (2/10) * (11 + 1) = 2.4, es decir, entre el segundo y el tercer dato. De nuevo, podemos tomar el promedio de estos dos valores para obtener D2.
3. Y así sucesivamente, hasta el noveno decil (D9).

Estos valores dividen el conjunto de datos en diez partes iguales en términos de número de observaciones.

```python
datos.Ingreso.quantile([i/10 for i in range(1,10)])
```

### **6.3 Percentiles**


Los percentiles son medidas estadísticas que dividen un conjunto de datos ordenados en cien partes iguales. Estas partes son conocidas como P1, P2, ..., P99. Cada percentil deja un cierto porcentaje de datos por debajo de él.

Para calcular los percentiles, primero ordenamos los datos de menor a mayor (o viceversa). Luego, encontramos la posición de cada percentil utilizando la fórmula:

* Para Pk: (k/100) * (n + 1)

Donde "k" es el número de percentil que estamos buscando (por ejemplo, k = 1 para P1, k = 50 para P50, k = 99 para P99), y "n" es el número total de datos en el conjunto.

Si el resultado de esta fórmula no es un número entero, se puede tomar el promedio de los valores en las posiciones inmediatamente superiores e inferiores para obtener el percentil correspondiente.

Por ejemplo, si tenemos los siguientes datos: 10, 15, 20, 25, 30, 35, 40, 45, 50, 55, 60.

1. El percentil 1 (P1) estaría en la posición (1/100) * (11 + 1) = 0.12, es decir, entre el primer y el segundo dato. Podemos tomar el promedio de estos dos valores para obtener P1.
2. El percentil 50 (P50) es la mediana, que estaría en la posición (50/100) * (11 + 1) = 6, es decir, en el sexto dato.
3. El percentil 99 (P99) estaría en la posición (99/100) * (11 + 1) = 11.88, es decir, entre el undécimo y el duodécimo dato. Podemos tomar el promedio de estos dos valores para obtener P99.

Estos valores dividen el conjunto de datos en cien partes iguales en términos de número de observaciones.

```python
datos.Ingreso.quantile([i/100 for i in range(1,100)])
```

**Grafico de distribucion de frecuencia acumulada**

Una gráfica de frecuencia acumulada es útil para visualizar la distribución acumulada de los datos en un conjunto de datos. En lugar de mostrar la frecuencia de cada valor individual, muestra cómo se acumulan los valores a lo largo de un rango de datos.

La gráfica de frecuencia acumulada puede proporcionar información sobre la dispersión de los datos, la concentración alrededor de ciertos valores y la variabilidad en los extremos del conjunto de datos. Además, permite una comparación visual rápida entre diferentes conjuntos de datos.

Algunos usos comunes de las gráficas de frecuencia acumulada incluyen:

1. Identificar los percentiles de los datos: La gráfica de frecuencia acumulada permite identificar rápidamente qué porcentaje de los datos está por debajo o por encima de cierto valor, lo que facilita la identificación de los percentiles.

2. Comparar distribuciones de datos: Al trazar múltiples gráficas de frecuencia acumulada en el mismo gráfico, se pueden comparar fácilmente las distribuciones de diferentes conjuntos de datos.

### **6.4 Box-plot**

Un boxplot, también conocido como diagrama de caja y bigotes, es una herramienta gráfica que se utiliza para representar la distribución de un conjunto de datos y para identificar posibles valores atípicos dentro de esos datos. Consiste en un rectángulo (la "caja") que abarca el rango intercuartílico (IQR por sus siglas en inglés), que es el rango entre el primer cuartil (Q1) y el tercer cuartil (Q3) del conjunto de datos. Dentro de la caja, una línea representa la mediana.

Los "bigotes" del boxplot se extienden desde los bordes de la caja hasta los valores extremos, que son los valores más extremos dentro de una cierta distancia (generalmente 1.5 veces el IQR) de los cuartiles. Los valores atípicos, que son aquellos que caen fuera de este rango, se representan como puntos individuales.

El boxplot proporciona una visualización compacta y efectiva de la distribución de los datos, permitiendo identificar fácilmente la mediana, el rango intercuartílico y los valores atípicos en el conjunto de datos. ## **7. Medidas de disérsion**


Las medidas de dispersión son estadísticas que describen la variabilidad o la dispersión de un conjunto de datos. Son útiles para comprender cuán dispersos están los valores individuales alrededor de la medida central, como la media o la mediana. Algunas de las medidas de dispersión más comunes incluyen:

1. **Desviación media absoluta:** Es la media de las diferencias absolutas entre cada valor y la media del conjunto de datos. Proporciona una medida de la dispersión promedio de los datos.

2. **Varianza:** Es la media de los cuadrados de las diferencias entre cada valor y la media del conjunto de datos. Proporciona una medida de la dispersión de los datos alrededor de la media, pero está sesgada por los valores cuadrados.

3. **Desviación estándar:** Es la raíz cuadrada de la varianza. Proporciona una medida de la dispersión de los datos en la misma escala que los datos originales, lo que facilita la interpretación.


### **7.1 Desviación media absoluta**


La desviación media absoluta (DMA) es una medida de dispersión que calcula el promedio de las diferencias absolutas entre cada valor de un conjunto de datos y la media de ese conjunto. Se calcula utilizando la siguiente fórmula:

dma.png

Para calcular la desviación media absoluta, primero se calcula la diferencia absoluta entre cada valor y la media del conjunto de datos. Luego, se toma el promedio de estas diferencias absolutas.

La desviación media absoluta proporciona una medida de la dispersión promedio de los datos alrededor de la media. Es menos sensible a los valores extremos en comparación con la desviación estándar, ya que no implica elevar al cuadrado las diferencias individuales.

Este estadístico es útil para comprender la variabilidad de los datos y es una alternativa robusta a la varianza y la desviación estándar cuando se tienen valores atípicos en el conjunto de datos.

### **7.2 Varianza**

La varianza es una medida de dispersión que describe cuánto varían los valores de un conjunto de datos en relación con la media de ese conjunto. Se calcula como la media de los cuadrados de las diferencias entre cada valor y la media del conjunto de datos. La fórmula para calcular la varianza es la siguiente:


varianza.png

Para calcular la varianza, primero se calcula la diferencia entre cada valor y la media del conjunto de datos. Luego, estas diferencias se elevan al cuadrado y se suman. Finalmente, se divide la suma de los cuadrados de las diferencias entre el número total de observaciones en el conjunto de datos para obtener la varianza.

La varianza es una medida útil para comprender la dispersión de los datos alrededor de la media. Sin embargo, debido a que implica elevar al cuadrado las diferencias individuales, está sesgada por los valores cuadrados y puede ser difícil de interpretar directamente. Por esta razón, a menudo se utiliza la desviación estándar, que es la raíz cuadrada de la varianza, ya que está en la misma escala que los datos originales y es más fácil de interpretar.
