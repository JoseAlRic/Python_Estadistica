* H0 = Las dos muestras de datos tiene la misma distribucion
* **H1 = Los valores de una muestra tienden a ser mayores que la otra**

* pValue < alpha --> No Ho

Si yo te muestro una moneda y te digo que voy a lanzarla para arriba, ¿cuál es la chance de que salga corona? Asumimos que es ½, partiendo del principio que la moneda no esté alterada.

Pero si lanzamos la moneda 3 veces para arriba y todas las veces dá cara, ya puedes comenzar a sospechar. Podrías hasta intentar argumentar: “sospecho que la moneda esté viciada para el lado de la cara, ya que los datos parecen indicar eso”.

En realidad 3 muestras es un número pequeño, pero, dejando ese punto de lado, hay algo extraño que está ocurriendo. Miramos los datos (3 veces cara) y llegamos a una hipótesis (la moneda está viciada para el lado de la cara). Ahí usamos nuevamente los datos de las observaciones para validar nuestra hipótesis.

Ese acto es llamado de double dipping, cuando se investiga dos veces el mismo dato, una para levantar una hipótesis y otra para validarla. Eso es peligroso, pues muchas veces la hipótesis será validada ya que se usaron los mismos datos para levantarla.

¿Cómo evitar esa situación? Basta validar la hipótesis en un conjunto diferente de datos, no con los mismos datos con que fue levantada. Esto significa, que tenemos que separar de antemano nuestros datos en más de una parte. Una parte para levantar hipótesis y modelos y otra para validarlos.
